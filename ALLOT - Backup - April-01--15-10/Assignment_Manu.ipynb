```python
# WAP to create a simple calculator.

def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Error! Division by zero."
    return x / y

print("Select operation:")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")

while True:
    choice = input("Enter choice (1/2/3/4): ")

    if choice in ('1', '2', '3', '4'):
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))

        if choice == '1':
            print("Result:", add(num1, num2))
        elif choice == '2':
            print("Result:", subtract(num1, num2))
        elif choice == '3':
            print("Result:", multiply(num1, num2))
        elif choice == '4':
            print("Result:", divide(num1, num2))
        
        break
    else:
        print("Invalid input")
```

```python
# WAP to shuffle and print a specified list.

import random

# Specify the list
my_list = [1, 2, 3, 4, 5]

# Shuffle the list using Fisher-Yates shuffle algorithm
for i in range(len(my_list) - 1, 0, -1):
    j = random.randint(0, i)
    my_list[i], my_list[j] = my_list[j], my_list[i]

# Print the shuffled list
print("Shuffled list:", my_list)
```

```python
# WAP to check whether a given number is palindrome or not using a for loop.

num = input("Enter a number: ")
is_palindrome = True

for i in range(len(num) // 2):
    if num[i] != num[-i - 1]:
        is_palindrome = False
        break

if is_palindrome:
    print(num, "is a palindrome")
else:
    print(num, "is not a palindrome")
```

```python
# WAP to swap the first and the last character of a string

def swap_first_last_char(input_string):
    if len(input_string) < 2:
        return input_string
    
    return input_string[-1] + input_string[1:-1] + input_string[0]

# Input string
input_string = input("Enter a string: ")

# Swap first and last character
swapped_string = swap_first_last_char(input_string)

# Print the swapped string
print("Original string:", input_string)
print("Swapped string:", swapped_string)
```

```python
# WAP to calculate the sum of three numbers, if the values are equal then return thrice of their sum using functions.

def sum_or_thrice_sum(a, b, c):
    total_sum = a + b + c
    if a == b == c:
        return total_sum * 3
    else:
        return total_sum

# Input three numbers
try:
    a = float(input("Enter first number: "))
    b = float(input("Enter second number: "))
    c = float(input("Enter third number: "))
except ValueError:
    print("Invalid input. Please enter valid numbers.")
    exit()

# Calculate sum or thrice sum based on input
result = sum_or_thrice_sum(a, b, c)

# Print the result
print(f'Result for {a} + {b} + {c}:  {result}')
```

```python
# WAP that creates an iterator that returns numbers indefinitely, starting with 1.

def number_generator():
    num = 1
    while True:
        yield num
        num += 1

# Create an iterator
iterator = number_generator()
i = int (input("Number till which the printing should be done till: "))
# Print the first 5 numbers from the iterator
for _ in range(i):
    print(next(iterator))
```

```python
# WAP to create a decorator to memoize functions with an arbitrary set of arguments.

from functools import wraps

def memoize(func):
    cache = {}
    @wraps(func)
    def wrapper(*args):
        if args in cache:
            return cache[args]
        result = func(*args)
        cache[args] = result
        return result
    return wrapper

@memoize
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

# Test the memoized fibonacci function
print(fibonacci(10))  # This will calculate fibonacci(10) and memoize the result
print(fibonacci(8))   # This will reuse the memoized result for fibonacci(8)
```

```python
# WAP to read a file and print the following
#  a) Number of characters (including whitespace)
#  b) Number of words (separated by whitespace)
#  c) Number of lines
#  d) Number of unique words (case sensitive)

def create_file(file_path):
    with open(file_path, 'w') as file:
        file.write("This is a sample text.\n")
        file.write("It contains words and sentences.\n")
        file.write("This is the third line.\n")
        file.write("Sample text for testing.\n")

def count_file_stats(file_path):
    num_chars = 0
    num_words = 0
    num_lines = 0
    unique_words = set()

    with open(file_path, 'r') as file:
        for line in file:
            num_lines += 1
            num_chars += len(line)
            words = line.split()
            num_words += len(words)
            unique_words.update(words)

    num_unique_words = len(unique_words)

    return num_chars, num_words, num_lines, num_unique_words

file_path = "example.txt"
create_file(file_path)
num_chars, num_words, num_lines, num_unique_words = count_file_stats(file_path)

print(f"Number of characters: {num_chars}")
print(f"Number of words: {num_words}")
print(f"Number of lines: {num_lines}")
print(f"Number of unique words: {num_unique_words}")
```

```python
# WAP that has a class Distance with members - kms and metres. Derive classes School and Office which store the distance from your house to school and Office which store the distance from your house to school and office along with other details.

class Distance:
    def __init__(self, kms, metres):
        self.kms = kms
        self.metres = metres

class School(Distance):
    def __init__(self, kms, metres, name, address):
        super().__init__(kms, metres)
        self.name = name
        self.address = address

class Office(Distance):
    def __init__(self, kms, metres, name, address):
        super().__init__(kms, metres)
        self.name = name
        self.address = address

# Example usage
school_distance = School(2, 200, "RKD Sasthamangalam", "Sasthamangalam")
office_distance = Office(16, 800, "Technopark", "Kazhakootam")

print(f"Distance from home to {school_distance.name}: {school_distance.kms} kms, {school_distance.metres} metres")
print(f"Distance from home to {office_distance.name}: {office_distance.kms} kms, {office_distance.metres} metres")
```

```python
# WAP to create two threads to keep a count of number of even numbers entered by the user.

import threading

# Global variable to store the count of even numbers
even_count = 0

# Maximum number of entries allowed
max_entries = 5

# Lock for synchronization
lock = threading.Lock()

# Function to keep track of even numbers
def count_even_numbers():
    global even_count
    global max_entries
    count = 0
    while count < max_entries:
        try:
            with lock:
                num = int(input("Enter a number (0 to exit): "))
        except ValueError:
            print("Invalid input. Please enter a valid number.")
            continue
        
        if num == 0:
            break
        if num % 2 == 0:
            even_count += 1
        count += 1

# Create two threads
thread1 = threading.Thread(target=count_even_numbers)
thread2 = threading.Thread(target=count_even_numbers)

# Start the threads
thread1.start()
thread2.start()

# Wait for the threads to finish
thread1.join()
thread2.join()

# Print the total count of even numbers
print(f"Total even numbers entered: {even_count}")
```

```python
# Concept of API's: WAP to implement a basic quiz application using Open Trivia API (https://opentdb.com/)

import requests
import json

def get_questions(amount=5, category=None, difficulty=None, type='multiple'):
    base_url = 'https://opentdb.com/api.php'
    params = {
        'amount': amount,
        'type': type
    }
    if category:
        params['category'] = category
    if difficulty:
        params['difficulty'] = difficulty

    response = requests.get(base_url, params=params)
    if response.status_code == 200:
        data = json.loads(response.text)
        return data['results']
    else:
        print('Failed to fetch questions')
        return []

def display_question(question, index):
    print(f"Question {index + 1}: {question['question']}")
    options = question['incorrect_answers'] + [question['correct_answer']]
    options = sorted(options) if question['type'] == 'multiple' else options
    for i, option in enumerate(options):
        print(f"{chr(65 + i)}) {option}")
    user_answer = input('Your answer (A/B/C/D): ')
    return user_answer.upper() == chr(65 + options.index(question['correct_answer']))

def main():
    print('Welcome to the Trivia Quiz!\n')
    questions = get_questions(amount=3, category=9, difficulty='easy')
    if not questions:
        print('No questions found.')
        return

    score = 0
    for index, question in enumerate(questions):
        print()
        correct = display_question(question, index)
        if correct:
            print('Correct!')
            score += 1
        else:
            print('Incorrect!')

    print(f'\nQuiz Complete! You scored {score}/{len(questions)}')

if __name__ == '__main__':
    main()
```

```python
# WAP to send an automated birthday wish on e-mail.

import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
import datetime

# Email configuration
SMTP_SERVER = 'smtp-mail.outlook.com'
SMTP_PORT = 587
sender_email = "vaishnavi.m@thinkpalm.com"
receiver_email = "manu.m@thinkpalm.com"
password = "Malu$123"

# Get the current date
current_date = datetime.date.today()

# Birthday date (replace with the recipient's birthday)
birthday_date = datetime.date(1985, 3, 25)

# Check if it's the recipient's birthday
if current_date.month == birthday_date.month and current_date.day == birthday_date.day:
    # Message configuration
    message = MIMEMultipart()
    message["From"] = sender_email
    message["To"] = receiver_email
    message["Subject"] = "Happy Birthday!"

    body = "Happy Birthday! Wishing you a day filled with joy and happiness."
    message.attach(MIMEText(body, "plain"))

    # Send the email
    with smtplib.SMTP(SMTP_SERVER, SMTP_PORT) as server:
        server.starttls()  # Enable encryption
        server.login(sender_email, password)  # Use your email address and password for authentication
        server.sendmail(sender_email, receiver_email, message.as_string())
    print("Birthday wish sent successfully!")
else:
    print("Today is not the recipient's birthday.")
```
